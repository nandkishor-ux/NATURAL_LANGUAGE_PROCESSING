```python
sentences="Elias stood at the corner of 5th Avenue, a striking figure amidst the morning rush. His silver hair, usually meticulously combed, was slightly disheveled by the morning wind, and his antique pocket watch was a constant companion. Known to neighbors as a quiet man, he actually spent hours restoring old, abandoned bicycles, giving them a second life. Elias often wore a tweed jacket, regardless of the season, and could be found whispering encouragement to his plants on his balcony. His kind, thoughtful demeanor masked a sharp wit that often surprised new acquaintances in the bustling neighbourhood"
```

```python
import nltk
words=nltk.word_tokenize(sentences)
```

```python
nltk.pos_tag(words)
```

```python
import nltk

# Download the correct chunker data
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')  # ✅ this is the one actually needed
nltk.download('words')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')

tag_elements = nltk.pos_tag(words)
w = nltk.ne_chunk(tag_elements)
print(w)
```
