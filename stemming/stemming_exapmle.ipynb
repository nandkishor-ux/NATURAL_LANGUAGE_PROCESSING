```python
words=["eating","sleeping","coding","running","swimming","dancing","singing","painting","writing","traveling","eats","sleeps","codes","runs","swims","dances","sings","paints","writes","travels","history"]
```

##PORTERSTEMMER

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for word in words:
    print(word,"--->",stemming.stem(word))
```

```python
stemming.stem("congratulation")
```

Regexpstemmer class: these are the classes with the help help of ehich we can easily imp;ement reguler . It basically takes a single word and remove any suffix or prefix


```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|able$' , min=3)
reg_stemmer.stem("eating")

```

Snowball Stemmer , so basically it is the upgraded version of all the libraries 


```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer("english")
for word in words:
    print(word,"--->",snowballstemmer.stem(word))
```
