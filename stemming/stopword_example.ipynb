```python
paragraph="""The old lighthouse stood at the edge of the rocky cliff, overlooking the vast blue ocean. Every night, its bright beam of light swept across the dark waters, guiding lost ships safely toward the shore.
The keeper of the lighthouse was a quiet man who loved his solitary life.
He would wake up early each morning, climb the tall spiral staircase, and polish the giant glass lens until it sparkled. Seagulls often circled around the tower, calling out loudly in the salty breeze.
Sometimes, during heavy storms, the waves would crash violently against the rocks below, but the lighthouse always stood firm.
 One day, a young sailor arrived at the lighthouse, seeking shelter from the raging storm.
 The keeper welcomed him warmly and offered him a place to stay until the storm passed. As they sat by the fire, the sailor shared stories of his adventures at sea, while the keeper listened intently, feeling a sense of camaraderie he hadn't experienced in years.
 The storm eventually subsided, and the sailor thanked the keeper for his hospitality before setting off on his next journey. The lighthouse continued to shine brightly, a beacon of hope and safety for all who ventured into the vast ocean.
 """
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
sentences=nltk.sent_tokenize(paragraph)
type(sentences)
```

```python
##apply stopword and filter and apply stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    sentences[i]=' '.join(words)
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
##apply stopword and filter and apply stemming
for i in range(len(sentences)): 
    words=nltk.word_tokenize(sentences[i]) 
    words=[snowballstemmer.stem(word) for word in words if word not in stopwords.words('english')] 
    sentences[i]=' '.join(words)
##apply stopword and filter and apply stemming
for i in range(len(sentences)): 
    words=nltk.word_tokenize(sentences[i]) 
    words=[snowballstemmer.stem(word) for word in words if word not in stopwords.words('english')] 
    sentences[i]=' '.join(words)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
##apply stopword and filter and apply stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word, pos='v') for word in words if word not in stopwords.words('english')]
    sentences[i]=' '.join(words)
```
