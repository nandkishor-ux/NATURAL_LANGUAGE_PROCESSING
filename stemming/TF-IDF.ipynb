```python
import pandas as pd

messages = pd.read_csv(r'D:\natural_language_processing\dataset\SMSSpamCollection.csv', 
                       encoding='utf-8')

messages.columns = ['index', 'label', 'message']  # file has 3 columns: #, Label, Message

print(messages.shape)
print(messages.head())
```

```python
messages
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatize=WordNetLemmatizer()
##from nltk.stem.porter import PorterStemmer
##ps=PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', str(messages['message'][i]))
    review = review.lower()
    review = review.split()
    review = [wordlemmatize.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

CREATE TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```
